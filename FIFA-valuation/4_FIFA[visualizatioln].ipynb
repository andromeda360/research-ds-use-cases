# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('FIFA1_EDA.csv')
```

```python
df
```

```python
df.drop(['Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control','Acceleration','Sprint Speed','Agility','Reactions','Balance','Shot Power','Jumping','Stamina','Strength','Long Shots','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Marking','Standing Tackle','Sliding Tackle','GK Diving', 'GK Handling','GK Kicking', 'GK Positioning', 'GK Reflexes','day','month','dayofweek','Total','ID','empty','end_year','loan'],axis=1,inplace=True)
```

```python
def position_type(p):
    if (p['Best_position']=='GK'):
        return 'G'
    elif ((p['Best_position']=='CB') | (p['Best_position']=='RB') | (p['Best_position']=='LB') | (p['Best_position']=='RWB') | (p['Best_position']=='LWB')):
        return 'D'
    elif ((p['Best_position']=='CM') | (p['Best_position']=='CDM') | (p['Best_position']=='CAM')| (p['Best_position']=='RM') | (p['Best_position']=='LM')):
        return 'M'
    elif ((p['Best_position']=='ST') | (p['Best_position']=='CF') | (p['Best_position']=='RW') | (p['Best_position']=='LW')):
        return 'F'
    else:
        return 0
    
df['Best_position'] = df.apply(position_type,axis = 1)
```

# Univariate Analysis

```python
columns = ['Age', 'overall_rating', 'potential','Height', 'Weight','foot', 'Best_overall', 'Best_position','Growth','week_foot','skill_move', 'Attacking_work_Rate', 'defensive_work_Rate','International_Reputation', 'Body Type']
for i in columns:
  fig, ax = plt.subplots(1,1, figsize=(13, 6))
  sns.countplot(x = df[i], palette='Blues_r')
```

```python
plt.figure(figsize=(15,35))
sns.countplot(data=df,y='national',order=df['national'].value_counts().index)
```

# Bivariate Analysis

```python
columns = ['Age', 'overall_rating', 'potential','Height', 'Weight','foot', 'Best_overall', 'Best_position','Growth','week_foot','skill_move', 'Attacking_work_Rate', 'defensive_work_Rate','International_Reputation', 'Body Type',]
for i in columns:
  fig, ax = plt.subplots(1,1, figsize=(17, 6))
  sns.barplot(y = 'Value',x = df[i][1:],data=df.iloc[1:], palette='Blues_r')
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x = 'Age',y = 'potential',data=df, palette='Blues_r')
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x = 'Best_position',y = 'Height',data=df, palette='Blues_r')
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x = 'Best_overall',y = 'Value',data=df, palette='Blues_r')
```

```python
sns.relplot(data=df,x='overall_rating',y='Value',palette='viridis',hue='Age',sizes=(15,200),aspect=2)
```

```python
sns.relplot(data=df,x='potential',y='Value',palette='viridis',hue='Age',sizes=(15,200),aspect=2)
plt.title('potential vs Value wrt age',fontsize=20)
```

```python
sns.relplot(data=df,x='potential',y='Value',palette='viridis',hue='Best_position',sizes=(15,200),aspect=2)
plt.title('potential vs Value wrt best position',fontsize=20)
```
